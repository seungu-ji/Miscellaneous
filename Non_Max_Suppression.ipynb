# **Non Max Suppression**

NMS(Non Max Suppression)이 무엇이고 왜 쓰이는지에 대해 알아보겠다.

image.png

위의 그림에서 왼쪽 부분을 보겠다. 그림을 보면 알 수 있듯이 object detection algorithm은 object가 있다고 추정되는 부분에 많은 detection을 하는 경향이 있다.

그래서 오른쪽 그림과 같이 **NMS**를 활용하여 detection이 된 object의 bounding box 중 비슷한 bounding box들을 제거하고 제일 적절한 bounding box만 남긴다.

## **NMS Algorithm**

이제 NMS의 수행 로직에 대해 자세히 알아보겠다.

1. Detected된 bounding box 별로 특정 Confidence threshold 이하 bounding box는 먼저 제거를 한다. (EX. confidence score < 0.5 => 0.5 미만을 모두 제거)
2. confidence score를 내림차순으로 정렬한다.
3. 정렬된 box를 높은 confidence socre를 가진 box와 겹치는 다른 box를 모두 조사하여 IOU가 특정 threshold 이상인 box를 모두 제거한다. (EX. IoU Threshold > 0.4 => IoU 점수가 0.4 초과인 것을 모두 제거)
4. 남아있는 bounding box들만 선택한다.

image.png

위의 이미지처럼 NMS을 거치면 하얀 bounding box만 남고 나머지 빨간 bounding box는 제거가 된다.

정리하자면, NMS는 여러 bounding box중 최상의 bounding box만 남게 해준다. Confidence score가 높을수록, 그리고 IoU Threshold가 낮을수록 bounding box들이 많이 제거가 된다.
