# **1-Stage Detector & 2-Stage Detector**

위의 그림에서 가운데 선을 기준으로 위에 있는 것들이 2-Stage Detector이며, 아래에 있는 것들이 1-Stage Detector이다.

## **2-Stage Detector**

image.png

2-Stage Detector의 경우 위의 그림과 같이 regional proposal과 classification이 순차적으로 이루어진다.

이때 regional proposal이란, image에서 object detection을 진행하기 위해 '물체가 있을만한 영역'을 찾는 것이다. 기존에는 **sliding window** 방식을 사용하였지만, selective search, edge boxes 등을 사용한다.

## **1-Stage Detector**

image.png

1-Stage Detector는 2-Stage Detector와 반대로 **regional proposal과 classification이 동시에 진행**된다. 즉, 2-Stage Detector에 비하면 빠르지만, 정확도는 낮다.
